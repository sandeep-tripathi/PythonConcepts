```
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
# Path of the file in google drive
filename = '/content/drive/MyDrive/fandisk.stl'

# Create a new plot
figure = pyplot.figure()
axes = mplot3d.Axes3D(figure)

# Loading fandisk.stl
your_mesh = mesh.Mesh.from_file(filename)

# Adding vector to the plot
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(your_mesh.vectors))

# View can be changed by changing parameter (90,270) with values 0,90,180 and 270
#axes.view_init(90,270)

#Auto scale to the mesh size
scale = your_mesh.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)
print('Visualization for the mesh object:')
#Show the plot to the screen
pyplot.show()

# Calculation of surface normal
# I assume it to be facet normal as documentation for library is not clear 
your_mesh_normal=your_mesh.normals
print("Calculation of surface normal:")
print(your_mesh_normal)

```
